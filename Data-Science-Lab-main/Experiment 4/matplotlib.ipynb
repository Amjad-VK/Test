```python
#dashed graph
from matplotlib import pyplot as pt
x=[1,2,3,4]
y=[1,3,5,7]
pt.plot(x,y,linewidth=3.0,linestyle="dashed")
pt.show()
```

```python
#dotted graph
from matplotlib import pyplot as pt
x=[1,2,3,4]
y=[1,3,5,7]
pt.plot(x,y,linewidth=2.0,linestyle="dotted")
pt.show()
```

```python
#Grid
from matplotlib import pyplot as pt
x=[1,2,3,4]
y=[1,3,5,7]
pt.grid()
pt.plot(x,y)
pt.show()
```

```python
#bar graph
from matplotlib import pyplot as pt
pt.bar([1,2,3],[1,3,2])
pt.show()
```

```python
#pie chart
from matplotlib import pyplot as pt
la=['Apple','banana','orange']
x=[2,4,6]
pt.pie(x,labels=la)
pt.show()
```

```python
#histogram
import numpy as np
from matplotlib import pyplot as pt
x=np.random.normal(100,2,250)
pt.hist(x)
pt.show()
```

```python
x=[1,4,7,20]
y=[2,5,8,11]
pt.subplot(5,5,1)
pt.plot(x,y)
pt.show()
x=[1,4,7,20]
y=[11,8,5,2]
pt.subplot(5,5,2)
pt.plot(x,y)
pt.show()
```

```python
from matplotlib import pyplot as pt
la=['Apple','banana','tomato']
x=[2,4,6]
y=[0.2,0,0]
pt.pie(x,labels=la,explode=y,shadow=True)
pt.show()
```

```python
#To plot a graph
from matplotlib import pyplot as plt
a=[1,2,3,4]
plt.plot(a)
plt.show()
```

```python
#Plot a graph with x and y array values
import numpy as np
from matplotlib import pyplot as pt
x=np.arange(0,10,0.1)
y=np.sin(x)
pt.plot(x,y)
pt.xlabel("x axis")
pt.ylabel("y axis")
pt.title("sin graph")
pt.show()
```

```python
pt.scatter(x,y)
pt.show()
```

```python
#To color
pt.plot(x,y,'b+')
pt.show()
```

```python
from matplotlib import pyplot as pt
x=[0,1,2,3,4,5,6,7]
y=[1,5,7,9,11,13,15,17]
pt.plot(x,'r')
pt.plot(y,'g*')
pt.xlabel("x axis")
pt.ylabel("y axis")
pt.title("Graph")
pt.legend(['x','y'])
pt.show()
```
