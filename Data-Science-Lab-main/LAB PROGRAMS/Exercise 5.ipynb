```python
import numpy as np
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as pt
% matplotlib inline
a=Image.open('a.png')
b=np.array(a)
c=ndimage.rotate(b,45,cval=120)
d=Image.fromarray(c)
pt.imshow(d)
```
