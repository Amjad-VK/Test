```python
#Simple Calculator
a=int(input("Enter First Number:"))
b=int(input("Enter Second Number:"))
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")
option=int(input("Choose an Option:"))
if(option==1):
    print("Sum:",a+b)
elif(option==2):
    print("Difference:",a-b)
elif(option==3):
    print("Product:",a*b)
elif(option==4):
    print("Division:",a/b)
else:
    print("Invalid")
```
