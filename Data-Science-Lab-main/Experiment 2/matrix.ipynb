```python
# Program To Perform Matrix Multipilcation And Transformation Using Python .

#Matrix Multiplication
import numpy as np
a=np.arange(4).reshape((2,2))
b=np.arange(4).reshape((2,2))
c=np.dot(a,b)
print(c)
```

```python
# Rotation
import numpy as np
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as pt
%matplotlib inline
a=Image.open('a.jpg')
b=np.array(a)
c=ndimage.rotate(b,45,cval=120)
d=Image.fromarray(c)
pt.imshow(d)
```

```python
# Scaling
import numpy as np
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as pt
%matplotlib inline
a=Image.open('a.jpg')
b=np.array(a)
c=ndimage.zoom(b,(0.5,0.5,1))
d=Image.fromarray(c)
pt.imshow(d)
```

```python
# Shearing
import numpy as np
from PIL import Image
from scipy import ndimage
from matplotlib import pyplot as pt
%matplotlib inline
a=Image.open('a.jpg')
b=np.array(a)
h,w,s=b.shape
m=[[1,0.5,0],[0.5,1,0],[0,0,1]]
c=ndimage.affine_transform(b,m,offset=(0,h//2,0),output_shape=(h,h+w//2,s))
d=Image.fromarray(c)
pt.imshow(d)
```
